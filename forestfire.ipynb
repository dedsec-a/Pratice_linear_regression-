```python
# Dividing the Features into Dependent and Independent Features 
x = df_copy.drop('FWI', axis= 1)
y = df['FWI']
```

```python
print(x.head() , y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,Y_train , Y_test = train_test_split(x, y , test_size = 0.3 , random_state= 42)
```

```python
print(X_train.head() ,Y_train.head())
print(X_train.shape , X_test.shape)
```

```python
# Feature Selection Based  on Co-realtion 
plt.figure(figsize=(20,10))
corr = X_train.corr()
sns.heatmap(corr , annot= True)
```

```python
# Now we will be Checking Multi-Colinearity 
# Here we are taking the Thersohold Randomly but the real threshold is set by the Domain Expert 

def corelation (dataset , thershold ) :
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range (i) :
            if abs(corr_matrix.iloc[i , j]) > thershold :
                col_name = corr_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr
```

```python
corr_features = corelation(X_train , 0.85)
```

```python
# Droping the Feartures --> Droping the features when it is more than the thershold Value 
X_test.drop(corr_features, axis= 1 ,inplace= True)
X_train.drop(corr_features,axis =1 ,inplace= True)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Feature Scaling and Stardarzation 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled  = sc.fit_transform(X_train)
X_test_scaled  = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , mean_absolute_error
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X_train_scaled , Y_train)
```

```python
y_preds = model.predict(X_test_scaled)
mae = mean_absolute_error(Y_test ,y_preds)
mse = mean_squared_error(Y_test , y_preds)
r2 = r2_score(Y_test , y_preds)
print("The mean Absoluute error is " , mae)
print("The mean squared Error is " , mse)
print("The R2 Score is " , r2)

```

```python
plt.scatter(Y_test , y_preds)
```

```python
# Lasso regresion model 
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error , mean_absolute_error
from sklearn.metrics import r2_score
model1 = Lasso()
model1.fit(X_train_scaled , Y_train)
y_preds1 = model1.predict(X_test_scaled)
mae_1 = mean_absolute_error(Y_test ,y_preds1)
mse_1 = mean_squared_error(Y_test ,y_preds1)
r2_1 = r2_score(Y_test ,y_preds1)
```

```python
print("The Mean absolute error for the Lasso reg is ", mae_1)
print(" the mean squared eroor for the lasso reg is ", mse_1)
print("the r2 score for the lasso reg is ", r2_1)
```

```python
plt.scatter(Y_test , y_preds1)
```

```python
# applying the Ridge Regression Model 
from sklearn.linear_model import Ridge
model2 = Ridge()
model2.fit(X_train_scaled , Y_train)
y2_preds = model2.predict(X_test_scaled)
mae_2 = mean_absolute_error(Y_test , y2_preds)
mse_2 =  mean_squared_error(Y_test , y2_preds)
r2_2 = r2_score(Y_test , y2_preds)
```

```python
print("The Mean Absolute error in the ridge model is" , mae_2 )
print("The Mean squared Error is " , mse_2)
print("The R2 Score of the model is ", r2_2)
```
